```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.random.rand(50)
y= np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.scatter(x,y,color = "orange",alpha = 0.4,marker = "X")
plt.title("Scatter plot")
plt.grid()
plt.figure(figsize=(6,2))
```
