```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip)
```

```python
tips["smoker"].value_counts()
```

```python
tips["smoker"].unique()
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip,data = tips,hue = "smoker")
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip , data = tips,hue = "sex")
```

```python
tips["size"].value_counts()
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,hue = "size")
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,style = "size")
```

```python
tips["day"].unique()
```

```python
tips["time"].unique()
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tips, hue = "time")
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tips, style = "day")
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tips, style = "size",hue = "time")
```

```python
sns.catplot(x = "day",y = "total_bill",data = tips)                # used to represent categorical data and numerical data
```

```python
sns.catplot(x = "smoker",y= "tip",data = tips)
```

```python
sns.jointplot(x = tips.total_bill,y = tips.tip)              # this is combination of histogram and scatter plot
```
