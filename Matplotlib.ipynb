#Matplotlib


* useful for making plots 

```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
x = np.linspace(-10,10,100)
```

```python
y = np.sin(x)
z = np.cos(x)
print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.show
plt.title("Sine")
```

```python
plt.plot(x,z)
plt.xlabel("angles")
plt.ylabel("cosine values")
plt.title("COSINE CURVE")
plt.show
```

```python
r = np.linspace(-10,10,50)
print(r)
```

```python
p = r*r
plt.plot(r,p,"r*")
plt.show
```

```python
x = np.linspace(-5,5,100)
plt.plot(x,np.sin(x),"r")
plt.plot(x,np.cos(x),"g--")
plt.show
```

```python
 fig = plt.figure()
 ax = fig.add_axes([0,0,1,1])
 language = ["english","french","hindi","german","spanish"]
 people = [1000,200,300,400,50]
 ax.bar(language,people)
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
language = ["english","french","hindi","german","spanish"]
people = [1000,200,300,400,50]
ax.pie(people,labels = language,autopct = '%1.1f%%')                  #added the %1.1f%% to add one floating point in the percentage representation of value
plt.show()
```

```python
x = np.linspace(0,10,30)
y =  np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = "g")
ax.scatter(x,z,color = "r")
ax
plt.show()
```

3D scatter plot


```python
fig3 = plt.figure()
ax = plt.axes(projection = "3d")
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap= "Oranges")
plt.show()
```
