#NUMPY

```python
import numpy as np
from time import process_time
```

```python
list1 = [i for i in range(100000000)]
start_time = process_time()

list1 = [i+5 for i in list1]
end_time = process_time()
print(end_time - start_time)
```

```python
np_array = np.array([i for i in range(100000000)])
start_time = process_time()
np_array += 5   
end_time = process_time()
print(end_time - start_time)
```

#initial placeholders in numpy array

```python
# if we need any particular initial value 
# example - create an array with zero 
x = np.zeros((4,5),dtype = int)
print(x)
```

```python
y= np.ones((4,5))
print(y)
```

```python
z = np.full((5,4),5)      # to create an array with any particular value (example here is 5)
print(z)
```

```python
a = np.eye(4,dtype = int)      #create an identity matrix
print(a)
```

```python
# create an array using random values 
r = np.random.random((3,4))
print(r)
# Here the values printed are between 1 and 0 
```

```python
# Create an array for random value between any two particular values
r = np.random.randint(10,100,(3,4))         #np.random.randint(low,high,size,dtype)
print(r)
```

```python
#array of evenly spaced numbers
d= np.linspace(10,30,5,dtype = int)             #np.linspace(start,stop,no_of_value)
print(d)
```

```python
#array of evenly spaced values --> specifying the step
e = np.arange(10,30,5)      #np.arange(start,stop,step)
print(e)
```

```python
#convert a list to an array
list2= [1,2,3,4,5]
arr =  np.asarray(list2)
print(arr)
type(arr)
```

#Mathematical operations on a Numpy array


```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
print(list1+list2)       # it concatenate or joins two list 
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a+b)
```

```python
a = np.random.randint(10,30,(3,3))
b = np.random.randint(20,30,(3,3))
print(a)
print(" ")
print(b)
print(" ")
print(a+b)
```

```python
print(a*b)     #it will perform the element wise multiplication
```

```python
print(a-b)
```
