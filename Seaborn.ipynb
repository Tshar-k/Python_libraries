Data visualisation library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

Seaborn has some builtin dataset 

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.tail()
```

```python
#setting theme for the plot 
sns.set_theme()
```

```python
sns.relplot(data = tips, x = "total_bill", y = "tip" , col = "time",hue = "smoker",style = "smoker",size = "size")
```

```python
#Load the iris dataset 
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",hue = "species",data= iris)
```

```python
sns.relplot(x="sepal_length",y="petal_length",hue = "species",data= iris)
```

```python
sns.scatterplot(x="sepal_length",y="petal_width",hue = "species",data= iris)
```

```python
sns.scatterplot(x="sepal_width",y="petal_width",hue = "species",data= iris)
```
